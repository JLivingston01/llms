```python
def make_convo(string, default_response):

    convos = [
        (i,default_response) for i in string.split("\n")
    ]
    return convos
```

```python
convos = sum([
    make_convo(weather_talk,'let me tell you about the weather.'),
    make_convo(schedule_talk,'Let me tell you about your calendar.'),
    make_convo(greeting_talk,'I am well. Can I help you today?'),
    make_convo(hello,'Hi! Can I help you today?'),
    make_convo(time,'The time is..'),
    make_convo(news,'Let me fetch the news.'),
    make_convo(messages,'Let me get that message ready to send.'),
    make_convo(searching,'Let me look that up for you.'),
    make_convo(questions,'Checking on that...'),
    ],start=[])
```

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import StackingClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier

from sklearn.preprocessing import LabelEncoder
```

```python
estimators = [
    ('rfc', RandomForestClassifier(random_state=42,max_depth=10,min_samples_leaf=2,verbose=True)),
    ('rfc2', RandomForestClassifier(random_state=42,max_depth=25,min_samples_leaf=1,verbose=True,)),
    ('rfc3', RandomForestClassifier(random_state=42,n_estimators=500,max_depth=None,min_samples_leaf=1,verbose=True,)),
    ('rfc4', RandomForestClassifier(random_state=42,n_estimators=250,max_depth=25,min_samples_leaf=1,verbose=True,)),
    ('knc', KNeighborsClassifier(n_neighbors=2,
            metric='cosine',weights='distance',)),
    ('svc', SVC(gamma='scale',kernel='rbf',C=.10,tol=1e-3,degree=3,verbose=True))
    ]
final_estimator = None

reg = VotingClassifier(
    estimators=estimators,
    voting='hard',
    weights = [1,1,1,1,1,1,1]
)

reg2 = StackingClassifier(
    estimators=estimators,
    final_estimator=MLPClassifier(random_state=42,activation='relu',learning_rate_init=.0001,learning_rate='constant',
                                 max_iter=20000,batch_size=64,hidden_layer_sizes=(50,),verbose=True,n_iter_no_change=150,
                                 validation_fraction=0,early_stopping=False,tol=1e-12),
)

model = Pipeline(
    steps=[
        ('vect',
         CountVectorizer(binary=True,dtype=int,ngram_range=(1,1),),
         #TfidfVectorizer(binary=True,norm=None,sublinear_tf=True,use_idf=True)
         ),
        ('learner',reg2)
    ]
)
```
